```python
import pandas as pd

df = pd.read_csv("d1.csv")
df.head()
```

```python
df = df[df['fault_type'].isin([0,1])]
df['fault_type'].unique()
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
df['timestamp'] = df['timestamp'].dt.time
df[['timestamp']].head()
```

```python
df = df.drop('bulb_number', axis=1)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['environmental_conditions'] = le.fit_transform(df['environmental_conditions'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
y = df['fault_type']

X = df.drop('fault_type', axis=1)
```

```python
X['timestamp'] = pd.to_datetime(X['timestamp'].astype(str)).dt.hour
```

```python
X['timestamp'].head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)

print(X_train.shape)
print(X_test.shape)
```

```python
clean_df = pd.concat([X, y], axis=1)

clean_df.to_csv("clean_streetlight_dataset.csv", index=False)
```
