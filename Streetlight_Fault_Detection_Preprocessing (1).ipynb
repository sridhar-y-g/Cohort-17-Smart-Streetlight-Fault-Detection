```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("street_light_fault_detection.csv")

df.head()
```

```python
df = df[df['fault_type'].isin([0,1])]

df['fault_type'].unique()
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])

df['timestamp'] = df['timestamp'].dt.strftime('%H:%M:%S')
```

```python
df[['timestamp']].head()
```

```python
df = df.drop('bulb_number', axis=1)

df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['environmental_conditions'] = le.fit_transform(df['environmental_conditions'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
y = df['fault_type']

X = df.drop('fault_type', axis=1)
```

```python
X_numeric = X.drop('timestamp', axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X_numeric)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)

print(X_train.shape)
print(X_test.shape)
```
